```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.spatial import cKDTree
from sklearn.neighbors import BallTree
from tqdm import tqdm

from seisnet.utils import (filter_nc_evts_by_distance, get_data_dir,
                           get_repo_dir, load_picks_within_radius)
```

California [CRS 3310](https://epsg.io/3310-1188)

```python
df10k = filter_nc_evts_by_distance(10000,return_df=True)
df5k = filter_nc_evts_by_distance(5000,return_df=True)
df1k = filter_nc_evts_by_distance(1000,return_df=True)
df500m = filter_nc_evts_by_distance(500,return_df=True)
```

```python
fig,axs = plt.subplots(2,2,figsize=(8,10), sharex=True, sharey=True, gridspec_kw={"hspace":0.1, "wspace":0.1})
axs = axs.flatten()

df10k.plot(markersize=0.1,ax=axs[0])
df5k.plot(markersize=0.1,ax=axs[1])
df1k.plot(markersize=0.1,ax=axs[2],alpha=0.3)
df500m.plot(markersize=0.1,ax=axs[3],alpha=0.3)

axs[0].set_title(f"Sep. dist. = 10km, N = {len(df10k)}")
axs[1].set_title(f"Sep. dist. = 5km, N = {len(df5k)}")
axs[2].set_title(f"Sep. dist. = 1km, N = {len(df1k)}")
axs[3].set_title(f"Sep. dist. = 0.5km, N = {len(df500m)}")

plt.savefig(f"{get_repo_dir()}/figures/Sparse_evt_dist.png",
                dpi=200,edgecolor="none",bbox_inches="tight")
plt.show()
```
